```python
from IPython.display import display
from IPython.html.widgets import interactive
import skimage as sk
import skimage.io
```

```python
d = sk.io.imread('AERONET_BSRN_BAO_Boulder.2014245.terra.721.250m.tif')
```

```python
def threshold_blue(b_t=255):
    rgb_yellow = [255, 255, 0]
    mask = d[:,:,2] >= b_t
    d_masked = d.copy()
    d_masked[mask,:] = rgb_yellow
    sk.io.imshow(d_masked)
    num_selected_pixels = mask.sum()
    percent_coverage = 100*float(num_selected_pixels)/float(d.size/3)
    display("Number of selected pixels: %d [%2d%%]" % (mask.sum(), percent_coverage))
    return mask
```

```python
sk.io.imshow(d)
```

```python
w = interactive(threshold_blue, b_t=[0, 255])
display(w)
```

```python
sk.io.imshow(w.result)
```
